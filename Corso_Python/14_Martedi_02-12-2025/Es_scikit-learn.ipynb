```python
import sys
print(sys.executable)

from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = load_iris()
df_iris = pd.DataFrame(data.data, columns=data.feature_names)
df_iris['species'] = pd.Categorical.from_codes(data.target, data.target_names)
```

```python
print("=== Statistiche descrittive ===")
print(df_iris.describe())
```

```python
print("\n=== Prime 5 righe ===")
print(df_iris.head())
```

```python
print("\n=== Tipi di dato delle colonne ===")
print(df_iris.dtypes)
```

```python
print("\n=== Valori mancanti per colonna ===")
print(df_iris.isnull().sum())
```

```python
print("\n=== Info Dataset ===")
print(df_iris.info())
```

```python

# 7Ô∏è‚É£ Scatter plot Petal Length vs Petal Width
colors = {'setosa':'red', 'versicolor':'green', 'virginica':'blue'}
plt.figure(figsize=(8,6))
for species in df_iris['species'].unique():
    subset = df_iris[df_iris['species'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], 
                label=species, color=colors[species], s=50)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width per specie')
plt.legend()
plt.show()
```

```python
# 8Ô∏è‚É£ Boxplot delle feature per specie
plt.figure(figsize=(12,8))
for i, feature in enumerate(data.feature_names):
    plt.subplot(2,2,i+1)
    data_to_plot = [df_iris[df_iris['species']==species][feature] for species in df_iris['species'].unique()]
    plt.boxplot(data_to_plot, labels=df_iris['species'].unique())
    plt.title(f'Distribuzione di {feature} per specie')
plt.tight_layout()
plt.show()
```

```python
# 9Ô∏è‚É£ Heatmap (correlazione) con Matplotlib
corr = df_iris.iloc[:,:4].corr()
plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Matrice di correlazione delle feature")
for i in range(len(corr)):
    for j in range(len(corr)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}", ha='center', va='center', color='black')
plt.show()
```
